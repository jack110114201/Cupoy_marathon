# 104人力銀行網站爬蟲練習
## 從104人力銀行網站爬取求職公司資訊。

```python
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.webdriver.chrome.options import Options
```

### 首先進入104人力銀行網站：https://www.104.com.tw/cust/list/index/
### 在"地區"及"產業"下拉式選單中，選擇你希望的選項，然後按下"找公司"按鈕。

### 開啟新網頁後，觀察新的網址內容，把網址複製下來。

```python
job_104_url = "https://www.104.com.tw/jobs/search/?area=6001001000&cat=2007001000&jobsource=2018indexpoc&ro=0"
```

### 開始爬取搜尋結果

```python
chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
```

```python


browser = webdriver.Chrome()
#browser = webdriver.Chrome(executable_path='../DAY18/chromedriver',chrome_options=chrome_options)
browser.get(job_104_url)  # 打開瀏覽器並連到網頁
time.sleep(2)  # delay一段時間等待網頁更新完成


html = browser.page_source
soup = BeautifulSoup(html, 'lxml')
company_list = soup.find("div", attrs={'id':"js-job-content"}).find_all("article", attrs={'class':"b-block--top-bord job-list-item b-clearfix js-job-item js-job-item--focus b-block--ad"})

while True:
    time.sleep(3)  # delay一段時間等待網頁更新完成
    #
    # 擷取公司名稱及簡介內容
    #
    for company in range(0,len(company_list)):
        # 因為內容太多，我們把爬取到的結果，寫入"company_list.txt"檔案中，稍後再來檢視
        company_name = company_list[company]['data-cust-name']
        f = open("company_list.txt", "a+", encoding='utf-8')
        f.write( company_name + "\n" )  # 寫入公司名稱
        try:
            company_desc = company_list[company].find("p", attrs={'class':"job-list-item__info b-clearfix b-content"}).text.replace("'","").replace('\n',"")
            f.write( company_desc + "\n" )  # 寫入公司簡介
            f.write( "--------------------------------------------------------------------------------" + "\n" )
            f.close()
            print(company_name)
            print(company_desc)
            print('--------------------------------------------------------------------------')
        except:
            print('--------------------------------------------------------------------------')
            continue
            
    # 點選下一頁按鈕，如果沒有代表到最後了
    try:
        search_btn = browser.find_element_by_class_name('js-next-page')
        search_btn.click()
        time.sleep(2)
        html = browser.page_source
        soup = BeautifulSoup(html, 'lxml')
        company_list = soup.find_all(class_='js-job-item')
        print("前往下一頁...", end=" ")
    except:
        print("===END===")
        break
    
```
