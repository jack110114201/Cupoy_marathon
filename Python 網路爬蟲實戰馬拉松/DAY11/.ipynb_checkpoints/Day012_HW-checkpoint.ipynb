### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, 'html.parser')

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text)
    print(d.find_all('em')[-1].text)
    print(d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
# 使用datafeame的方式儲存

import pandas as pd

date = []
title = []
classify = []

for d in soup.find(class_="part_list_2").find_all('h3'):
    date.append(d.find(class_="date").text)
    classify.append(d.find_all('em')[-1].text)
    title.append(d.find_all('a')[-1].text)
    

article_dict = {'classify':classify,'date':date,'title':title}

df = pd.DataFrame(article_dict).sort_values(by='classify',ascending = False).reset_index(drop=True)
df
```

### ③ 哪一個類別的文章最多

```python
count_dic = {}

for d in soup.find(class_="part_list_2").find_all('h3'):
    if (d.find_all('em')[-1].text) not in count_dic:
        count_dic[(d.find_all('em')[-1].text)]=1
    else:
        count_dic[(d.find_all('em')[-1].text)]+=1

print("文章最多的類別：",max(count_dic,key=count_dic.get))
```
