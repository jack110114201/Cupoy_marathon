### 比較一下範例檔案中的「r.text」與「json.loads(r.text)」讀出來的內容有什麼差異

```python
import requests
r = requests.get('https://api.github.com/events')
r.text
```

```python
import json
json.loads(r.text)
```

#### Ans : r.text 以純文字表示，json.loads則以巢狀結構呈現
