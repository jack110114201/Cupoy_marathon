## 思考一:請問NLP與NLU之間的關係與差異?

#### NLP與NLU關係:
- 自然語言處理(NLP) 需要透過自然語言理解(NLU)與自然語言生成(NLG)而產生。

#### NLP與NLU差異:
- NLP是讓電腦學會理解人類的語言邏輯，以閱讀、翻譯、解譯、理解的流程了解人類的思維的程序。
- NLU是涉及將人類語言分解為機器可讀格式，目的在於資料的預處理，能夠協助後續的統計機率方法、機器學習、深度學習等模型訓練、預測等任務。NLU比NLP更進階，可以進行語義萃取、情緒分析、摘要、對話機器人等進階分析應用。

## 思考二: 影片中的自然語言處理的步驟清楚嗎? 是否可列出?

1. 分詞 (segmentation、Tokenizing)
2. 停用詞 (stop words)
3. 詞幹提取 (stemming)'
4. 詞形還原 (Lemmatization)
5. 詞性標註 (speech tagging)
6. 命名實體標註 (named entity tagging)

## 思考三: NLP與ML之間的關聯性為何?

#### NLP 可以分為兩組不同的方法: 
- 第一種方法是基於規則的方法，涉及人工或策劃的規則集。 基於規則的方法尋找語言術語，例如“愛”、“恨”、“喜歡”和“不喜歡”。 肯定詞和否定詞的存在決定了一個句子是肯定的還是否定的。 
- 第二種方法使用使用機器學習的統計技術來開發算法模型。 例如，通過機器學習，數據科學家可以訓練一種算法來理解基於大數據集的情緒，微調模型以預測全新句子中的情緒。
